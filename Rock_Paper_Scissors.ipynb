```python
import random
import time
```

```python
rock = 1
paper = 2
scissors = 3
```

```python
name = {
    rock: "Rock",
    paper: "Paper",
    scissors: "Scissors"
}

rules = {
    rock: scissors,
    paper: rock,
    scissors: paper
}
```

```python
players_score = 0
computers_score = 0
```

```python
def start():
  print('We can play a game of Rock, Paper, Scissors')

  while game():
    pass
  scores()
```

```python
def game():
  player = move()
  computer = random.randint(1, 3)
  print(f'You chose {name[player]}')
  print(f'Computer chose {name[computer]}')
  result(player, computer)
  return play_again()
```

```python
def move():
  while True:
    player = input('Rock = 1\nPaper = 2\nScissors = 3\n ')
    try:
      player = int(player)
      if player in (1, 2, 3):
        return player
    except:
      print('Invalid Input')
```

```python
def result(player, computer):
  print('......')
  time.sleep(0.5)

  print('Computer throw {0}'.format(name[computer]))
  global players_score
  global computers_score

  if player == computer:
    print("Its a tie")

  else:
    if rules[player] == computer:
      print("You won")
      players_score +=1

    else:
      print("Computer won")
      computers_score +=1
```

```python
def play_again():
  answer = input("Do you wish to play again??? (y/n)")
  if answer in ('y', 'Y', 'yes', 'Yes'):
    return answer
  else:
    print("Thank you for playing")
```

```python
def scores():
  global players_score
  global computers_score
  print(f'Final Scores...')
  print(f'Player: {players_score} Computer: {computers_score}')
```

```python
if __name__ == '__main__':
  start()
```
