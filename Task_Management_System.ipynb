```python
# Features:
# Add a New task
# List Down all tasks
# Mark a test as a Complete
# Delete a task
# List out all the completed task
# Exit the Program
```

```python
import datetime
```

```python
tasks=[]
def add_tasks():
  title = input('Enter task title: ')
  description = input('Enter task description')
  due_date = input('Enter due date (YYYY-MM-DD)')
  due_date = datetime.datetime.strptime(due_date,'%Y-%m-%d')


  task = {
    'title': title,
    'description': description,
    'due_date': due_date,
    'completed': False
    }

  tasks.append(task)
  print('Task added succesfully')
  print(f'Task {title} added')

```

```python
# List Down all tasks

def list_tasks():
  if not tasks:
    print("No tasks found")
  else:
    print("All tasks are:")
    for i, task in enumerate(tasks):
      status = 'completed' if task['completed'] else 'Not completed'
      print(f"{i+1}. {task['title']} - {status} Due: {task['due_date'].strftime('%Y-%m-%d')} ")
```

```python
# Mark a test as a Complete
def mark_complete():
  list_tasks()

  task_number = int(input("Enter the task number to mark as complete: "))
  if 1<= task_number <= len(tasks):
    tasks[task_number-1]['completed'] = True
    print(f"Task {tasks[task_number - 1]['title']} completed")

  else:
    print("Invalid task number")
```

```python
# Delete a task

def delete_task():
  list_tasks()
  task_number = int(input("Enter the task number to delete "))
  if 1 <= task_number <= len(tasks):
    removed_tasks = tasks.pop(task_number -1)
    print(f"Task {removed_tasks['title']} deleted successfully")

  else:
    print("Invalid task number")
```

```python
# List out all the completed task

def list_completed_tasks():
  completed_tasks = []
  for task in tasks:
    if task['completed']:
      completed_tasks.append(task)
  if not completed_tasks:
        print("No completed tasks found")

  else:
    print("Completed tasks are: ")
    for task in completed_tasks:
      print(f"{task['title']} , Due_date = {task['due_date'].strftime('%Y-%m-%d')} ")

```

```python
def main():
  print("TASK MANAGEMENT SYSTEM")

  while True:
    print("1. Add a Task")
    print("2. List Down all Tasks")
    print("3. Mark a test as Complete")
    print("4. Delete a Task")
    print("5. List down all the completed tasks")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
      add_tasks()

    elif choice =='2':
      list_tasks()

    elif choice == '3':
      mark_complete()

    elif choice =='4':
      delete_task()

    elif  choice =='5':
      list_completed_tasks()

    elif choice == '6':
      print('Task Management System Exited')
      break

    else:
      print("Invalid choice. Please enter 1 - 6 ")

```

```python
if __name__ == '__main__':
  main()
```
